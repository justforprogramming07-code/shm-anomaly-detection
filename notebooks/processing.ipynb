```python
import pandas as pd
df = pd.read_csv(r'E:\SHM_ML\learning\data\1st_test\2003.10.22.12.09.13', sep = '\t', header=None)
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

col = df[0]

plt.figure(figsize=(12, 4))
plt.plot(col.values, linewidth=0.5)
plt.axhline(col.mean(), color='red', label='mean')
plt.axhline(col.mean() + 2*col.std(), color='orange', linestyle='--', label='+2 std')
plt.axhline(col.mean() - 2*col.std(), color='orange', linestyle='--', label='-2 std')
plt.axhline(col.mean() + 3*col.std(), color='green', linestyle='--', label='+3 std')
plt.axhline(col.mean() - 3*col.std(), color='green', linestyle='--', label='-3 std')
plt.legend()
plt.title('Sensor 0 - Acceleration Readings')
plt.show()
```
